```python
import numpy as np
from scipy.stats import norm 

```

```python
mu = 70
sigma = 8
n = 100
x_bar = 73
alpha = 0.02
```

```python
z_critical = norm.ppf(1-alpha/2)
z_critical
```

```python
z = (x_bar-mu)/(sigma/np.sqrt(n))
z 
```

```python
if abs(z) > z_critical:
    print("Reject null hypothesis")
else:
    print("Accept null hypothesis")
```
