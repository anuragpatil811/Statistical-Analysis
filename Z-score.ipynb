```python
import numpy as np 
```

```python
transaction = [100, 120, 130, 110, 115, 500]
```

```python
#STANDARD
s = np.std(transaction)
s
```

```python
#MEAN
u = np.mean(transaction)
u
```

```python
#Z-score for 100
z = (100-u)/s  
print("Z-score for 100:", z)
```

```python
#z-score for 120
z = (120-u)/s    
print("Z-score for 120:", z) 
```

```python
#z-score for 130
z = (130-u)/s    
print("Z-score for 130:", z) 
```

```python
from scipy.stats import zscore
```

```python
zscore(transaction)
```
