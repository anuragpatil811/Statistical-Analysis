```python
l = [12, 14, 15, 3, 18, 19, 100]
```

```python
new_l = np.sort(l)
new_l
```

```python
#INTERQUARTILE RANGE
```

```python
Q1 = np.percentile(new_l, 25)
Q1 
```

```python
Q3 = np.percentile(new_l, 75)
Q3 
```

```python
Q2 = np.percentile(new_l, 50)
Q2 
```

```python
IQR = Q3 - Q1 
IQR 
```

```python
max_val = Q3 + 1.5*IQR
max_val
```

```python
min_val = Q1 - 1.5*IQR 
min_val 
```
