```python
import numpy as np
from scipy.stats import norm
```

```python
mu = 168
x_bar = 169.5
sigma = 3.9
n = 36
alpha = 0.05
```

```python
z = (x_bar - mu)/(sigma/n)
z_critical = norm.ppf(1-alpha/2)
print("Z statistic:", z)
print("Critical Z values:", -z_critical, "to", z_critical)
```

```python
if abs(z) > z_critical:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```
